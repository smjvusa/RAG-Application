```python
# Document loaders - load data from website 'ADDRESS' build an application acts like a chatbot
loader = WebBaseLoader("https://www.fullstackexperts.in/")
```

```python
data = loader.load()
data
```

```python
text=data[0].page_content
text
```

```python
#indexing:Text splitiing
!pip install -U langchain-text-splitters
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter,CharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
texts = text_splitter
```

```python
texts=text_splitter.create_documents([text])
texts
```

```python
len(texts)
```

```python
text_splitter = CharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200)
text_splitter
```

```python
texts=text_splitter.create_documents([text])
texts
```

```python
len(texts)
```

```python
#Embedding and storing in vector databases
!pip install -qU langchain-chroma
```

```python
from langchain_chroma import Chroma
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
# Gemini Embeddings
embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")
embeddings
```

```python
vector_store = Chroma.from_documents(
    documents=texts,
    embedding = embeddings,
    persist_directory="/contents/vectordata"
)
vector_store
```

```python
#to look the embeddings
vector_store.get()
```

```python
#to see the particular document embeddings
doc1_emb = vector_store.get('0838081a-da17-4d15-bfaa-af74d18b171f', include = ['embeddings'])
doc1_emb
```

```python
#Retrieval
#Sentence transformer embeddings-384,OpenAI embedding-1536,Gemini Embeddings-3072
question = 'What are the courses offered by fullstack academy?'
```

```python
#it is not runnable
similar_docs = vector_store.similarity_search(question, k = 5)
similar_docs
```

```python
# Retreiver Runnable
ret = vector_store.as_retriever(search_type = 'similarity')
ret
```

```python
ret.invoke(question)
```

```python
#Generation
from langchain_core.prompts import ChatPromptTemplate, HumanMessagePromptTemplate, SystemMessagePromptTemplate
```

```python
sys_temp = 'Answer the user queries concisely'
hum_temp = '''
Based on the context provided, answer the user query effectievely
{context}

The user query is: {question}
'''
```

```python
prompt_template = ChatPromptTemplate.from_messages([
    SystemMessagePromptTemplate.from_template(sys_temp),
    HumanMessagePromptTemplate.from_template(hum_temp)
])
prompt_template
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough, RunnableParallel
```

```python
chain = {'context': ret, 'question': RunnablePassthrough()}|prompt_template|llm|StrOutputParser()
chain
```

```python
print(chain.invoke(question))
```
